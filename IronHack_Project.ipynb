```python

def docTable(id,token): 
  url='https://api.emploi-store.fr/partenaire/infotravail/v1/package_show?id='+id
  API_URL=url
  token='Bearer '+token
  headers={'Authorization': token}
  response = requests.post(API_URL,headers=headers)
  json=response.json()
  results= json['result']['resources']
  InfoRequest={}
  listField=['name','description','id']
  for field in listField:
        listValue=[]
        for j in results:
          listValue.append(j[field])
        InfoRequest[field]= listValue
  return InfoRequest

objetDoc=docTable('a4f9e4dd-365e-4542-839c-a93a2448e388',token)
df = pd.DataFrame(objetDoc)
df.style.set_caption('Top 10 Fields of Research by Aggregated Funding Amount')
df
```
